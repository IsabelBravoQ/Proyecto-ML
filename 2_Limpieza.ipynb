## 2.- SOLUCIÃ“N A LOS DIFERENTES TIPOS DE ESCALAS DE MAGNITUD EN EL DATASET DE TERREMOTOS.

En el dataset de terremotos, existe un problema con las escalas de las magnitudes.

Para todas las escalas, la magnitud es un valor adimensional, pero no es intercambiable entre dichas escalas.

&nbsp;

Las escalas presentes en el dataset son:

* Mw (Moment Magnitude): estÃ¡ndar actual, mÃ¡s consistente para eventos grandes.

* Ms (Surface wave magnitude): basada en ondas superficiales.

* Mb (Body wave magnitude): ondas de cuerpo, Ãºtil para distancias largas.

* ML (Local magnitude o Richter original): ondas locales.  

&nbsp;

Algunas de las ecuaciones empÃ­ricas de conversiÃ³n son:

| De â†’ A       | FÃ³rmula aproximada                      | Referencia       |
| ------------ | --------------------------------------- | ---------------- |
| **Ms â†’ Mw**  | Mw â‰ˆ 0.67 Ã— Ms + 2.07                   | Hanks & Kanamori |
| **Mb â†’ Mw**  | Mw â‰ˆ 0.85 Ã— Mb + 1.03                   | Scordilis, 2006  |
| **Ml â†’ Mw**  | Mw â‰ˆ 0.85 Ã— Ml + 1.03                   | EstimaciÃ³n comÃºn |
| **Mfa â†’ Mw** | Aproximar igual que Ms (si no hay otra) |                  |


**Nota:** Estas relaciones son **aproximaciones empÃ­ricas**, las conversiones pueden no ser precisas.




```python
# Vamos a crear una nueva columna que se llama magnitude_Mw aplicando las fÃ³rmulas anteriores a nuestro dataset de terremotos.
# En la nueva columna obtenida, eliminamos las filas en las que el dato falte.

def to_mw(row):
    if not pd.isna(row["eqMagMw"]):
        return row["eqMagMw"]
    elif not pd.isna(row["eqMagMs"]):
        return 0.67 * row["eqMagMs"] + 2.07
    elif not pd.isna(row["eqMagMb"]):
        return 0.85 * row["eqMagMb"] + 1.03
    elif not pd.isna(row["eqMagMl"]):
        return 0.85 * row["eqMagMl"] + 1.03
    elif not pd.isna(row["eqMagMfa"]):
        return 0.67 * row["eqMagMfa"] + 2.07
    elif not pd.isna(row["eqMagnitude"]):
        return row["eqMagnitude"]  # Si no hay otra, usar la general
    else:
        return np.nan

earthquakes["magnitude_Mw"] = earthquakes.apply(to_mw, axis=1)

earthquakes= earthquakes.dropna(subset=["magnitude_Mw"])

# Creamos tambiÃ©n una columna donde nos indique la escala de magnitud de donde partÃ­amos, si hay algÃºn dato nulo, elimino esa fila.
def get_mag_source(row):
    if pd.notnull(row["eqMagMw"]): return "Mw"
    elif pd.notnull(row["eqMagMs"]): return "Msâ†’Mw"
    elif pd.notnull(row["eqMagMb"]): return "Mbâ†’Mw"
    elif pd.notnull(row["eqMagMl"]): return "ML"
    return "None"

earthquakes["MagnitudeSource"] = earthquakes.apply(get_mag_source, axis=1)
earthquakes= earthquakes.dropna(subset=["MagnitudeSource"])

# Elimino las columnas iniciales de magnitudes, ya que con las dos columnas nuevas tenemos los datos que necesitamos.

mag_columns = [col for col in earthquakes.columns if col.startswith("eqMag") or col == "eqMagnitude"]

earthquakes = earthquakes.drop(columns=mag_columns)
```

## 3.- FUSIÃ“N DE LOS DOS DATASETS

Antes de seguir con el tratamiento de datos faltantes, unimos los dos datasets.

Nos quedamos con las siguientes colunmas del dataset de tsunamis (para rellenar huecos en los datos de los terremotos y con interÃ©s informativo aunque podrÃ­a ser interesante para futuros estudios de protecciÃ³n frente a tsunamis):
* 'earthquakeEventId'   
* 'oceanicTsunami'  
* 'maxWaterHeight'  
* 'eqDepth',
* 'deathsAmountOrderTotal',
* 'housesDestroyedTotal',
* 'damageMillionsDollarsTotal'
* 'regionCode'
* 'latitude'
* 'longitude'


Unimos los dos datasets mediante el id de los terremotos.

Queremos estudiar si a partir de un terremoto, tuvo ocurrencia un tsunami, asÃ­ que los datos faltantes en oceanicTsunami, los imputamos como 0 (no tsunamis) y convertimos esta columna en binaria (serÃ¡ nuestro target)

```python
tsunamis.columns
```

```python
colum_ts = [
    'earthquakeEventId', 
    'oceanicTsunami', 
    'maxWaterHeight', 
    'eqDepth',
    'deathsAmountOrderTotal',
    'housesDestroyedTotal',
    'damageMillionsDollarsTotal',
    'regionCode', 
    'latitude',
    'longitude'
]

dfmerge = pd.merge(earthquakes,tsunamis[colum_ts],left_on='id',right_on='earthquakeEventId',how='left')

# NaN en oceanicTsunami â†’ 0 (no tsunami)
dfmerge['oceanicTsunami'] = dfmerge['oceanicTsunami'].fillna(0).astype(int)

```

```python
dfmerge.info()
```

```python
#Voy a rellenar los datos faltantes del dataset de terremotos con las columnas repetidas seleccionadas en el dataset de tsunamis.
#EliminarÃ© las filas del dataset resultante en las que eqDepth siga sin tener dato.

cols_repetidas = [
    'eqDepth',
    'deathsAmountOrderTotal',
    'housesDestroyedTotal',
    'damageMillionsDollarsTotal'
]

for col in cols_repetidas:
    col_x = f"{col}_x"
    col_y = f"{col}_y"
    
    if col_x in dfmerge.columns and col_y in dfmerge.columns:
        dfmerge[col_x] = dfmerge[col_x].fillna(dfmerge[col_y])
        dfmerge = dfmerge.rename(columns={col_x: col})
        dfmerge = dfmerge.drop(columns=[col_y])

dfmerge = dfmerge.dropna(subset=['eqDepth']).reset_index(drop=True)

dfmerge[cols_repetidas].info()
```
